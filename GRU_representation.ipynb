```python
def read_data_pfam(fname, select_cols = None):
    """
    Helper function to load protein sequences. 
    """
    f = open(fname).read().strip().split("\n")

    col_names = f[0].split(",")

    # Substract one for the header
    n_seqs = len(f) - 1

    # Loop through sequences adding each one as a column to the df
    annot_seqs = [f[i].split(",") for i in range(1, n_seqs)]

    # Initialize pandas dataframe with sequences
    df = pd.DataFrame(annot_seqs, columns=col_names)

    if select_cols is not None:
        df = df[select_cols]

    return df


def initialize_network_weights(net, method = 'kaiming'): 
    """
    Initialize fully connected and convolutional layers' weights
    using the Kaiming (He) or Xavier method. 
    This method is recommended for ReLU / SELU based activations.
    """

    torch.manual_seed(4)

    if method == 'kaiming': 
        for module in net.modules():

            if isinstance(module, nn.Linear): 
                nn.init.kaiming_uniform_(module.weight)
                nn.init.uniform_(module.bias)

    elif method == 'xavier':
        for module in net.modules():

            if isinstance(module, nn.Linear): 
                nn.init.kaiming_uniform_(module.weight)
                nn.init.uniform_(module.bias)

    else: 
        print('Method not found. Only valid for kaiming or xavier initialization.')

    return net
```

### Define RNN model 

All right. We're ready to write down a simple RNN model. The first thing that comes to mind is how to map the sequences to real numbers. One way we could do it is using one-hot encoding. The problem with one-hot-encoding is that it makes the data sparse and non-informative. Because we have a relatively small vocabulary (20 aminoacids) it wouldn't be that much of a problem, but this could get really bad for other type of NLP tasks. Luckily there's an alternative : embedding layers. You can think of an embedding layer as (literally) a look-up table that maps a set of indices to the real numbers. They are similar in nature to fully connected layers, in the sense that they are trainable and are learning a "weight "matrix, however embedding layers do not perform matrix multiplications between the input and the weight matrix. They are literally a indexable weight matrix([read more here](https://stackoverflow.com/questions/47868265/what-is-the-difference-between-an-embedding-layer-and-a-dense-layer)).

In this sense they will map single indices to $\mathbb{R}^d$ where $d$ is the embedding dimension. This will be the input to our recurrent net. We will be using a GRU cell for the RNN.

The mathematical model for a single GRU cell is the following. 

Let $t$ be the current timestep of sequence model, the input $\mathbf{X}_t \in \mathbb{R}^{n \times d}$ and the hidden state of the last timestep is  $\mathbf{H}_{t-1} \in \mathbb{R}^{n \times h}$ where $n$ is the number of samples in the minibatch (in our case we'll work with $n$ =1 for now), $d$ is the dimension of the input (in our case the dimensionality of the embedding layer) in the sequence, and $h$ the number of hidden units. Then the reset gate $\mathbf{R}_{t}$ and update gate $\mathbf{Z}_t$ are computed as follows:


$$
\mathbf{R}_t =  \sigma \left( \mathbf{X}_t \mathbf{W}_{xr} + \mathbf{H}_{t-1} \mathbf{W}_{hr} + \mathbf{b}_r \right)\\[1em]
\mathbf{Z}_{t} = \sigma \left( \mathbf{X}_t \mathbf{W}_{xz} + \mathbf{H}_{t-1} \mathbf{W}_{hz} + \mathbf{b}_z \right)
$$

where the $\mathbf{W}_{xi} \in \mathbb{R}^{d \times h}$  and $\mathbf{W}_{hi} \in \mathbb{R}^{h\times h}$ are weight matrices and  $\mathbf{b}_i \in \mathbb{R}^{1 \times h}$ are biases. We use a sigmoid function to squash valued in the interval $(0,1)$. Thus the reset and update gates will live in $\mathbb{R}^{n \times h}$.

**Now, to take into account the influence of the reset gate on the hidden state, we can pointwise multiply with $\mathbf{R}_t$ with the previous hidden state**. Whenever the entries in the reset gate $\mathbf{R}_t$ are close to 1, we recover the conventional RNN. This leads to the following update : 

$$
\tilde{\mathbf{H}}_t = \text{tanh} \left( \mathbf{X}_t \mathbf{W}_{xh} + (\mathbf{H}_{t-1} \odot \mathbf{R}_t )\mathbf{W}_{hh}  + \mathbf{b}_h \right)
$$

where $\odot$ is the Haddamard product, i.e. pointwise multiplication.

Essentially, the update gate works as a weighted sum of the previous hidden state $\mathbf{H}_{t-1}$ and the new candidate state $\tilde{\mathbf{H}}_{t}$, where the weights are given by the individual elements of the update gate $\mathbf{Z}_t$. More formally the definitive update equation for a GRU is given by : 


$$
\mathbf{H}_t = \mathbf{Z}_t \odot \mathbf{H}_{t-1} + (1 - \mathbf{Z}_t) \odot \tilde{\mathbf{H}}_t
$$

Let's wrap up the concepts: **Whenever the update gate $\mathbf{Z}_t$ is close to 1, we simply retain the old state, conversely when it's close to zero, we simply retain the candidate hidden state $\tilde{\mathbf{H}}_t$**. When the update gate is close to 1 then, we essentially get to ignore the information from the new input $\mathbf{X}_t$. This design can help us cope with the vanishing gradient problem and better capture dependencies for time series with large timestep distances. 

Operationally, `torch.GRU` expects a tensor of inputs $X_t$ of size `(seq_len, batch_size, input_size)`. We will be working with tensors of size `(seq_len, 1, embedding_size)` for a given protein of `seq_len` aminoacids. `torch` allows for using minibatches by using padding, but we'll focus on that later on. `torch.GRU` returns an `output` tensor corresponding to all of the hidden states $H_t$ to $H_{t + \text{seq_len}}$ and a `h_n` tensor corresponding to the last hidden state $H_{t + \text{seq_len}}$. The output tensor will be the input for a fully connected layer that will map a sequence of hidden states to a sequence of aminoacids, i.e. a map from $\mathbb{R}^{\text{seq_len} \times h} \mapsto \mathbb{R}^{\text{seq_len} \times 21}$ . We will be using `LogSoftmax()` for the final layer for numerical stability (it uses Log-sum-exp trick [see this torch discussion](https://discuss.pytorch.org/t/logsoftmax-vs-softmax/21386)). 

A summary of the computational graph goes something like this: 

    * list_of_indices, shape (len(seq), 1) => embeddings, shape (len(seq),embedding_dim) => sequence_of_hiddens, shape (len(seq), hidden_dim) => output, shape (len(seq), 21) > log_softmax(output)

All right! Let's write down our model. **We will also include a helper method for projecting single aminoacids to the embedding space.**

```python
class RNN_GRU(nn.Module): 
    """
    Generative or multi-class classifier RNN using the GRU cell.
    
    Params 
    ------
    input_size (int):
        Vocabulary size for initializing the embedding layer. 
        In the case for single aminoacid prediction it is 20 (or 21).
        
    embedding_size (int):
        Dimensionality of the embedding layer. 
    
    hidden_size (int):
        Dimensionality of the hidden state of the GRU. 
        
    output_size (int): 
        Dimensionality of the output. If it's in the generative mode, 
        it is the same as the input size. 
    
    input (torch.LongTensor):
        Sequence of indices.
    
    Returns
    -------
    output (torch.tensor): 
        Sequence of predictions. In the generative case it is a probability
        distribution over tokens. 
    
    last_hidden(torch.tensor): 
        Last hidden state of the GRU. 
        
    """
    
    def __init__(self, input_size, embedding_size, hidden_size,
                 output_size, n_layers = 1):
        
        super(RNN_GRU, self).__init__()
        
        self.input_size = input_size 
        self.embedding_size = embedding_size
        self.hidden_size = hidden_size 
        self.output_size = output_size
        self.n_layers = n_layers 
        self.log_softmax = nn.LogSoftmax(dim= 1)
        
        
        self.embedding = nn.Embedding(input_size, embedding_size)
        
        self.gru = nn.GRU(
            input_size = embedding_size,
            hidden_size= hidden_size,
            num_layers = n_layers
        )
    
        self.decoder = nn.Linear(hidden_size, output_size)

        
    def aa_encoder(self, input): 
        "Helper function to map single aminoacids to the embedding space."
        projected = self.embedding(input)
        return projected 
    
    def forward(self, input): 
        
        embedding_tensor = self.embedding(input)
        
        # shape(seq_len = len(sequence), batch_size = 1, input_size = -1)
        embedding_tensor = embedding_tensor.view(len(input), 1, self.embedding_size)

        sequence_of_hiddens, last_hidden = self.gru(embedding_tensor)
        output_rnn = sequence_of_hiddens.view(len(input), self.hidden_size)

        output = self.decoder(output_rnn)
        # LogSoftmax the output for numerical stability
        output = self.log_softmax(output)
        return output, last_hidden
```

Now, let's define  a helper function to make one forward pass for the training.

```python
def train(rnn_model, input_tensor, target_tensor, loss_fn, optimizer):
    "Helper function to make a forward pass through the RNN model."
    
    # Zero out gradients 
    rnn_model.zero_grad()
    loss = 0
    
    # Make forward computation
    output, hidden = rnn_model(input_tensor)

    loss += loss_fn(output, target_tensor)
    
    # Backprop error and update params
    loss.backward()
    optimizer.step()
    
    # Return average loss
    return loss.item() / len(input_tensor)
```

Now let's initialize the network parameters, the loss function and optimizer. 

```python
embedding_size= 10
hidden_size = 128
n_aminoacids = len(aminoacid_list) + 1

rnn = RNN_GRU(
    input_size = n_aminoacids, 
    embedding_size = embedding_size, 
    hidden_size = hidden_size, 
    output_size = n_aminoacids
)

# Initialize fully connected layer with Xavier method 
rnn = initialize_network_weights(rnn, method = 'xavier')

n_epochs = int(0.8*len(df_dev))
print_every = 1000
n_layers = 1
learning_rate = 1e-3

criterion = nn.NLLLoss()

optimizer = optim.Adam(rnn.parameters(), lr = learning_rate)
```

### Extracting the hidden state of the GRU 

Okay great, now we have a trained RNN that can generate proteins. Now we want to make a function to represent our protein sequence. We will take a similar approach to that of [Alley et al.](https://www.nature.com/articles/s41592-019-0598-1), by taking the last hidden state of the GRU after passing the protein through the trained RNN. **The idea is that the hidden state encodes information about the history of the RNN. By extracting the last hidden state we can get a snapshot of how the network *felt* right after generating that protein**.

```python
def get_hidden_state(sequence, rnn_model, hidden_size, token_to_ix, 
                     get_last_hidden = True): 
    """
    Extract the last or average hidden state of a given sequence. 
    """
    
    with torch.no_grad():
        sequence_indices = Variable(torch.LongTensor(token_to_ix(sequence)))
        
        # Initialize hidden state with random normal vector
        #hidden = Variable(torch.randn(1, 1, hidden_size))
        
        # Forward pass
        all_hidden, last_hidden = rnn_model(sequence_indices)
        
        if get_last_hidden: 
            return last_hidden.view(hidden_size).numpy()
        else:   
            return all_hidden.mean(axis = 0)
```

### Visualizing proteins on hidden state space. 

A simple visualization is to take all of the proteins' hidden state projection, and then apply a simple PCA on them. A PCA is a technique for dimensionality reduction that is based on getting the directions of highest variance in a dataset. [Here is a beautiful explanation through visualization of the technique](https://setosa.io/ev/principal-component-analysis/).  

Because PCA is a rotation, the first two PCs will convey a linear combination of the two first dimensions of maximal variance in the dataset. In contrast to other nonlinear embedding methods like tSNE, we could be over confident if we see clustering of the protein families. With PCA we can be confident that we'll have a good and unbiased visualization of our hidden state space.
